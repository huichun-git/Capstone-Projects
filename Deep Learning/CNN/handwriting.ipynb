```python
import pandas as pd
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/semeion/semeion.data",delim_whitespace=True, header=None,)
```

```python
import numpy as np
```

```python
df1 = df.iloc[:,:256]
df2 = df.iloc[:,256:]
```

```python
df1.head()
```

```python
df2.head()
```

```python
g = df1.values   #convert pandas to numpy
```

```python
g=g.reshape(1593,16,16,1)
```

```python
h = df2.values
```

```python
h.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dropout
from keras.layers import MaxPooling2D
from keras.layers import Conv2D, Flatten, Dense, Dropout
```

```python
batch_size = 100 #https://stats.stackexchange.com/questions/153531/what-is-batch-size-in-neural-network
num_classes = 10
epochs = 30
img_rows, img_cols = 16,16
input_shape = (img_rows, img_cols,1)  #define input shape to add the input layer

model = Sequential()
model.add(Conv2D(32, kernel_size=(8, 8),activation='relu',input_shape=input_shape, padding='SAME'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(4, 4)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax')) 
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer="Nadam",
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(g, h, test_size=0.5, random_state=42)
#x_train,y_train = g,h
```

```python
y_train
```

```python
history=model.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
         validation_data=(X_test, y_test))
```

```python
import h5py
model.save('model1.h5')
del model  # deletes the existing model
```
